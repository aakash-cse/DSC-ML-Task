### Exploratory Data Analysis (EDA), also known as Data Exploration, is a step in the Data Analysis Process, where a number of techniques are used to better understand the dataset being used.

__‘Understanding the dataset'__
* Extracting important variables and leaving behind useless variables
* Identifying outliers, missing values, or human error
* Understanding the relationship(s), or lack of, between variables
* Ultimately, maximizing your insights of a dataset and minimizing potential error that may occur later in the process

# Lets separate the textual,categorical and numerical features in the dataset

```python
text_cols = ['How would you describe your ethnicity?',
             'Please describe what free transport means to you in a word, or words?',
             'Where else would you visit less than you do now, if you had to pay for travel?',
            'If you lose free travel, would this change the college or school you are considering attending?']
from wordcloud import WordCloud

for c in text_cols:
    print(len(data[c].unique()))
```

```python
text = data[text_cols[0]].values 

wordcloud = WordCloud().generate(str(text))

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

#### From this we come to people mindset about the ethnicity is some kind of racisim like white black and continental like Asian... etc

```python
text = data[text_cols[1]].values 

wordcloud = WordCloud().generate(str(text))

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

#### It describes like most of the words like freedom, financial, family.... people kind of words are repeated. which is obvious if we are using some govt or private resources which is free
